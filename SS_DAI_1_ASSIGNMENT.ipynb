```python
df['release_date'] = pd.to_datetime(df['released_year'].astype(str) + '-' +
                                    df['released_month'].astype(str) + '-' +
                                    df['released_day'].astype(str))

#droppin the original year, month, and day columns
df.drop(columns=['released_year', 'released_month', 'released_day'], inplace=True)
```

```python
#Remove unncessary columns
df.drop(columns=['bpm','key','cover_url',] , inplace=True)
df.drop_duplicates(subset=["track_name"] , inplace=True)
df.fillna(method="ffill",inplace=True)
df.reset_index(drop=True , inplace=True)
```

```python
# df['streams'] = df['streams'].astype('int') is generating an error because one value is string
```

```python
df = df[df['streams'].apply(lambda x: str(x).isdigit())]

df.reset_index(drop=True, inplace=True)
```

```python
df.shape
```

```python
df.streams = df.streams.astype(int)
```

```python
df.streams.dtype
```

```python
df["mode"].value_counts().plot(kind="pie",autopct="%1.1f%%")
plt.show()
```

```python
#top 10 most streamed artists on Spotify
plt.figure(figsize=(8,5))

plt.title('10 most streamed artists on Spotify' )
sns.barplot(df.artist_name.value_counts().head(10),palette='plasma')
plt.yticks(np.arange(6,38,2))
plt.xlabel('Most Streamed Artists')
plt.xticks(rotation=45)
plt.ylabel('Number of Songs streamed')
plt.tight_layout()
plt.show()
```

```python
#Songs by Artist Count
plt.bar(df["artist_count"].value_counts().index,df["artist_count"].value_counts().values)
plt.xlabel("Number of Artists")
plt.ylabel("Songs Count")
plt.title('Artist Count on a Single Song')
plt.show()
```

```python
#No. of hits in each of the past 5 years
df['release_date'].dt.year.value_counts().head(5).sort_index().plot(kind='bar')
plt.title('#Hits in last 5 years')
plt.xlabel('Year')
plt.ylabel('#Hits')
plt.show()
```

## **Multivariate Analysis**

```python
# Sorting the DataFrame by 'streams' in descending order and selecting the top 10 rows
top_10_streamed = df.sort_values(by='streams', ascending=False).head(10)[['track_name', 'streams']]

plt.figure(figsize=(10,8))
plt.bar(top_10_streamed['track_name'], top_10_streamed['streams'],color='red')

plt.xticks(rotation=45, ha='right')
plt.ylabel('Streams')
plt.xlabel('Track Name')
plt.title('Top 10 Most Streamed Songs')

plt.tight_layout()
plt.show()
```

```python
#Relationship Between Most Streamed Songs And Spotify Playlists In 2023
plt.figure(figsize = (20, 10))

sns.regplot(x = 'streams', y = 'in_spotify_playlists', data = df, scatter = True, fit_reg = True)

plt.title('Relationship Between Most Streamed Songs And Spotify Playlists In 2023')
plt.xlabel('Streams')
plt.ylabel('Number of Spotify Playlists')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
#Relationship Between Most Streamed Songs And Apple Playlists In 2023
plt.figure(figsize = (20, 10))

sns.regplot(x = 'streams', y = 'in_apple_playlists', data = df , scatter = True, fit_reg = True)

plt.title('Relationship Between Most Streamed Songs And Apple Playlists In 2023')
plt.xlabel('Streams')
plt.ylabel('Number of Apple Playlists')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
#10 Songs w/c occured most in Spotify Charts
top_10_spotify_charts = df.sort_values(by='in_spotify_charts', ascending=False).head(10)[['track_name', 'in_spotify_charts']]

plt.figure(figsize=(10,10))
plt.bar(top_10_spotify_charts['track_name'], top_10_spotify_charts['in_spotify_charts'],color='blue')

plt.xticks(rotation=45, ha='right')
plt.ylabel('Times in Spotify Charts')
plt.xlabel('Track Name')
plt.yticks(np.arange(80,160,5))
plt.ylim(80,160)
plt.title('10 Songs w/c occured most in Spotify Charts')

plt.tight_layout()
plt.show()
```

```python
#10 Songs w/c occured most in Apple Charts
top_10_apple_charts = df.sort_values(by='in_apple_charts', ascending=False).head(10)[['track_name', 'in_apple_charts']]

plt.figure(figsize=(10,10))
plt.bar(top_10_apple_charts['track_name'], top_10_apple_charts['in_apple_charts'],color='blue')

plt.xticks(rotation=45, ha='right')
plt.ylabel('Times in Apple Charts')
plt.xlabel('Track Name')
plt.yticks(np.arange(200,300,5))
plt.ylim(200,300)
plt.title('10 Songs w/c occured most in Apple Charts')

plt.tight_layout()
plt.show()
```

```python
sns.boxplot(data=df, x='mode', y='energy_%')
plt.title('Energy Distribution by Mode')
plt.show()
```

```python
sns.pairplot(df, vars=['danceability_%', 'energy_%', 'valence_%', 'acousticness_%'], hue='mode')
plt.show()
```

```python
#Relationship between All audio features

plt.figure(figsize = (12, 8))
corr = df[['danceability_%','valence_%', 'energy_%', 'acousticness_%', 'instrumentalness_%','liveness_%', 'speechiness_%']].corr()
sns.heatmap(corr, annot = True, cmap = 'coolwarm', vmin =- 1, vmax = 1)
plt.title('Correlation Heatmap of Audio Features')
plt.xticks(rotation = 45)
plt.yticks(rotation = 45)
plt.tight_layout()
plt.show()
```
